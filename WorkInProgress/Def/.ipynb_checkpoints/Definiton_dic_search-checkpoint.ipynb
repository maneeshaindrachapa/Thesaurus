```python
word_def_list = []
word_list = []
for i in range(100):
    temp = []
    word = synonyms.loc[i]['Word']
    print (word)
    temp.append(word)
    definition = synonyms.loc[i]['Def1']
    temp.append(definition)
    if (word!='' and definition!=''):
        word_list.append(word)
        word_def_list.append(temp)
```
